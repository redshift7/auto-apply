```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from autots import AutoTS


df = pd.read_csv("""C:/Users/sathv/Desktop/UNH/Capstone/w10/Dataset_Time_Series_I_LAB_BikeRentals.csv""")

model = AutoTS(
forecast_length=14, # Time Horizon. Number of future periods to forecast
frequency='infer', # AutoTS will infer the frequency of the date column like 'D' for daily
prediction_interval=0.95, # Confidence interval
ensemble=['simple', 'horizontal-min'], # Use ensemble methods
model_list='superfast', # the model will output a list of models very fast
max_generations=5, # Number of generations for the genetic algorithm
num_validations=2, # Number of validations
no_negatives=True, # all forecasted values should be greater than zero (positive)
)


model = model.fit(df, date_col='date', value_col='cnt')

prediction = model.predict()
forecast_values = prediction.forecast
print(forecast_values)




model.export_template("Sathvik_Kasoju_section_model.csv", #export the models to this csv file. Write your first name, last name, and section number
models="best", #export the best models
max_per_model_class=1, #export 1 model per model family (e.g., ARIMA, Prophet, XGBoos)
include_results=True, #include training and cross validation results in the csv file
)
```
